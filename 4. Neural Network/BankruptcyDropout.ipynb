```python
import pandas as pd
df = pd.read_csv(r"G:\Ddrive\PG DBDA\12 Practical Machine Learning_\Cases\Bankruptcy\Bankruptcy.csv")
df.head()
```

```python
X = df.iloc[:,2:]
X.head()
```

```python
y = df.iloc[:,1]
y.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.3, 
                                                    random_state=202,stratify=y)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Activation, Input
from tensorflow.keras import metrics
```

```python
tf.random.set_seed(2024)
input_layer = Input(shape = X_train.shape[1])
dense_1 = tf.keras.layers.Dense(20, activation='relu',activity_regularizer = keras.regularizers.l1(0.001),name="Dense_1")(input_layer)
dense_2 = tf.keras.layers.Dense(10, activation='relu',activity_regularizer = keras.regularizers.l1(0.001), name = "Dense_2")(dense_1)
dense_3 = tf.keras.layers.Dense(5, activation='relu',activity_regularizer = keras.regularizers.l1(0.001), name = "Dense_3")(dense_2)
output_layer = tf.keras.layers.Dense(1, activation='sigmoid', name = "output")(dense_3)
model = Model(inputs = input_layer, outputs = output_layer)

print(model.summary())
'''Dense_layer1 = Dense(10, input_dim=X_train.shape[1], activation='relu',
                     activity_regularizer = keras.regularizers.l1(0.001))
Dense_layer2 = Dense(8, activation='relu',
                     activity_regularizer = keras.regularizers.l1(0.001))
Dense_layer3 = Dense(4, activation='relu',
                     activity_regularizer = keras.regularizers.l1(0.001))
Dense_layer4 = Dense(1,activation='sigmoid')'''
```

```python
model.compile(optimizer='sgd', loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit( X_train,y_train,validation_data=(X_test,y_test),verbose=2,epochs=200)
```

```python
y_pred_prob = model.predict(X_test)
y_pred_prob[:5]
```

```python
from sklearn.metrics import log_loss
log_loss(y_true=y_test,y_pred=y_pred_prob)
```

```python
from keras.layers import Dropout
```

```python
tf.random.set_seed(seed=2024)
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(20, activation='relu',activity_regularizer = keras.regularizers.l1(0.001),input_shape=(X_train.shape[1], )), 
    tf.keras.layers.Dropout(rate=0.2,seed=2024),
    tf.keras.layers.Dense(10, activation='relu',activity_regularizer = keras.regularizers.l1(0.001)),
    tf.keras.layers.Dropout(rate=0.1,seed=2024),
    tf.keras.layers.Dense(5, activation='relu',activity_regularizer = keras.regularizers.l1(0.001)), 
    tf.keras.layers.Dense(1, activation='sigmoid')  
])
model.compile(optimizer='sgd', loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit( X_train,y_train,validation_data=(X_test,y_test),verbose=2,epochs=200)
```

```python
from sklearn.metrics import log_loss
y_pred_prob = model.predict(X_test)
log_loss(y_true=y_test,y_pred=y_pred_prob)
```
